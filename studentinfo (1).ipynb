```python
import pandas as pd

# Load the dataset
file_path = '/content/studentInfo.csv'
df = pd.read_csv(file_path)

# Get summary of the dataset
summary = df.describe(include='all')

# Get the data types of the variables
data_types = df.dtypes

# Get the count of missing and null values
missing_values = df.isnull().sum()



print("\nData Types of the Variables:")
print(data_types)

print("\nMissing Values Count:")
print(missing_values)
```

```python
# Impute missing values in 'imd_band' with the mode
mode_imd_band = df['imd_band'].mode()[0]
df['imd_band'].fillna(mode_imd_band, inplace=True)
```

```python
df= df.dropna()
```

```python
# Get the count of missing and null values
missing_values = df.isnull().sum()
print("\nMissing Values Count:")
print(missing_values)
```

```python
import matplotlib.pyplot as plt

# Count the occurrences of each category in the 'final_result' column
result_counts = df['final_result'].value_counts()

# Create a bar plot for the 'final_result' column
plt.figure(figsize=(10, 6))
result_counts.plot(kind='bar', color=['green', 'orange', 'red', 'gold'])
plt.title('Final Result Distribution')
plt.xlabel('Final Result')
plt.ylabel('Number of Students')
plt.xticks(rotation=0)
plt.grid(axis='y')

# Display the plot
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt



# Define the function to plot individual columns and their relationship with 'final_result'
def plot_column_vs_final_result(column_name):
    fig, axs = plt.subplots(1, 2, figsize=(15, 5))

    # Plot individual column distribution
    df[column_name].value_counts().plot(kind='bar', ax=axs[0], color='skyblue')
    axs[0].set_title(f'{column_name} Distribution')
    axs[0].set_xlabel(column_name)
    axs[0].set_ylabel('Number of Students')
    axs[0].tick_params(axis='x', rotation=90)

    # Plot column vs 'final_result'
    pd.crosstab(df[column_name], df['final_result']).plot(kind='bar', stacked=True, ax=axs[1])
    axs[1].set_title(f'{column_name} vs Final Result')
    axs[1].set_xlabel(column_name)
    axs[1].set_ylabel('Number of Students')
    axs[1].tick_params(axis='x', rotation=90)

    plt.tight_layout()
    plt.show()

# Plot for 'code_module' and 'code_presentation'
columns_to_plot = ['code_module', 'code_presentation']
for column in columns_to_plot:
    plot_column_vs_final_result(column)
```

```python
# Plot for 'gender' and 'region'
columns_to_plot = ['gender', 'region']
for column in columns_to_plot:
    plot_column_vs_final_result(column)
```

```python
# Plot for 'highest_education' and 'imd_band'
columns_to_plot = ['highest_education', 'imd_band']
for column in columns_to_plot:
    plot_column_vs_final_result(column)
```

```python
# Plot for 'age_band', 'num_of_prev_attempts', 'studied_credits', and 'disability'
columns_to_plot = ['age_band', 'num_of_prev_attempts', 'disability']
for column in columns_to_plot:
    plot_column_vs_final_result(column)
```

The dataset contains the following columns:

1. **code_module**: Encoded value representing the module code.
2. **code_presentation**: Encoded value representing the presentation code.
3. **id_student**: Unique identifier for each student.
4. **gender**: Encoded value representing gender (1 for male, 0 for female).
5. **region**: Encoded value representing the student's region.
6. **highest_education**: Encoded value representing the highest level of education attained.
7. **imd_band**: Index of Multiple Deprivation band (socioeconomic status indicator).
8. **age_band**: Encoded value representing the age group of the student.
9. **num_of_prev_attempts**: Number of previous attempts at the module.
10. **studied_credits**: Number of credits the student is studying.
11. **disability**: Encoded value representing disability status (1 for yes, 0 for no).
12. **final_result**: Encoded value representing the final result (e.g., pass, fail).

```python
import pandas as pd
import numpy as np

# Assuming final_merged_df_with_result is your main dataset
# Create a copy of the original dataset
sorted_df = final_merged_df_with_result.copy()


# Step 2: Ensure the data is sorted by student and  date
sorted_df.sort_values(by=['id_student', 'date_submitted'], inplace=True)




sorted_df.head(50)
```

```python
# calculate days_to_deadline
sorted_df['days_to_deadline'] = sorted_df['date'] - sorted_df['date_submitted']

# calculate interaction terms
# Example interaction term: Interaction between days_to_deadline and score
sorted_df['interaction_registration_score'] = sorted_df['days_to_deadline'] * sorted_df['score']

# calculate days_between_submissions
sorted_df['days_between_submissions'] = sorted_df.groupby('id_student')['date_submitted'].diff()

# calculate submission_speed
sorted_df['submission_speed'] = sorted_df['date_submitted'] - sorted_df['date']

# Final Exam Score: Use score where assessment_type is 'Exam'
sorted_df['final_exam_score'] = sorted_df.apply(
    lambda row: row['score'] if row['assessment_type'] == 'Exam' else None, axis=1)
```

```python
import pandas as pd

# Assuming final_merged_df_with_result is your main dataset

# Function to assign phases based on date and module_presentation_length
def assign_phase(row):
    course_length = row['module_presentation_length']
    phase_25 = course_length * 0.25
    phase_50 = course_length * 0.50
    phase_75 = course_length * 0.75

    if row['date'] <= phase_25:
        return 'phase_25'
    elif row['date'] <= phase_50:
        return 'phase_50'
    elif row['date'] <= phase_75:
        return 'phase_75'
    else:
        return 'phase_100'

# Apply the function to assign phases
final_merged_df_with_result['phase'] = final_merged_df_with_result.apply(assign_phase, axis=1)

final_merged_df_with_result['phase'].value_counts()
```

```python
# Function to assign phases based on date and module_presentation_length
def assign_phase(row):
    course_length = row['module_presentation_length']
    phase_25 = course_length * 0.25
    phase_50 = course_length * 0.50
    phase_75 = course_length * 0.75

    if row['date_submitted'] <= phase_25:
        return 'phase_25'
    elif row['date_submitted'] <= phase_50:
        return 'phase_50'
    elif row['date_submitted'] <= phase_75:
        return 'phase_75'
    else:
        return 'phase_100'

# Apply the function to assign phases
sorted_df['phase'] = sorted_df.apply(assign_phase, axis=1)

# Verify the phase assignment
print(sorted_df['phase'].value_counts())
```

```python
# Calculate features for each phase
phase_features = sorted_df.groupby(['id_student', 'code_module', 'code_presentation', 'phase']).agg({
    'score': ['mean', 'min', 'max', 'std'],
    'submission_speed': 'mean',
    'days_between_submissions': 'mean',
    'is_banked': 'sum',
    'id_assessment': 'count',  # Count the number of submissions in each phase
    'days_to_deadline': 'mean',
    'interaction_registration_score': 'mean',
    'final_exam_score': 'mean'
}).reset_index()

# Flatten multi-level columns
phase_features.columns = ['_'.join(col).strip() for col in phase_features.columns.values]

# Rename columns for clarity
phase_features.rename(columns={
    'id_student_': 'id_student',
    'code_module_': 'code_module',
    'code_presentation_': 'code_presentation',
    'phase_': 'phase',
    'score_mean': 'avg_score',
    'score_min': 'min_score',
    'score_max': 'max_score',
    'score_std': 'std_score',
    'submission_speed_mean': 'avg_submission_speed',
    'days_between_submissions_mean': 'avg_days_between_submissions',
    'is_banked_sum': 'total_banked',
    'id_assessment_count': 'num_submissions',
    'days_to_deadline_mean': 'avg_days_to_deadline',
    'interaction_registration_score_mean': 'avg_interaction_registration_score',
    'final_exam_score_mean': 'avg_final_exam_score'
}, inplace=True)

# Verify the features
phase_features.head(20)
```
